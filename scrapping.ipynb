### News Scrapping

```python
from requests_html import HTMLSession,PyQuery as pq
session=HTMLSession()
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

import numpy as np
from datetime import datetime,timedelta
import json
```

```python
sid.polarity_scores('At least nine people were killed in the first strike when missiles destroyed a moving vehicle in the North Waziristan tribal region, the officials said.')
```

```python
start_date=datetime(2011,1,1)
end_date=datetime.now().date()
dt=timedelta(days=1)
for i in np.arange(start_date,end_date,dt).astype(datetime):
    print(type(str(i.date())))
    break
```

```python
#change it to save date in required use case format and change saving json format to [{'a':'b'},{'a':'b'}...]
f=open('headlines.json','w',encoding='utf-8')
all_data={}
for i in np.arange(start_date,end_date,dt).astype(datetime):
    
    while 1:
        temp=f'https://www.dawn.com/archive/{str(i.date())}'
        r=session.get(temp)

        #finds all articles in page
        articles=r.html.find('article')
        
        if len(articles)>0:
            break
            
        print("in while")
        
    print(len(articles))
    all_data[str(i.date())]=''
    for article in articles:
        #convert article to pyquery object
        t=pq(article.html)
        
#         #get news link
#         news_link=t('.theme--pakistan').siblings('h2').find('a').attr('href')
#         #read that separate news page
#         r2=session.get(news_link)
#         #read headline
#         text=r2.html.find('.caption')[0].text.split(' ')[0]
        
        #get only headings from articles/main page
        heading_text=t('.theme--pakistan').siblings('h2').text()
        if heading_text:
#             print(heading_text)
#             json.dump({str(i.date()):heading_text},f)
#             f.write(str(i.date())+","+heading_text+"\n")
            all_data[str(i.date())] += heading_text + '.'
    
#     if str(i.date()) == '2011-01-12':
#         break
    
json.dump(all_data,f,ensure_ascii=False)
    
f.close()
```
